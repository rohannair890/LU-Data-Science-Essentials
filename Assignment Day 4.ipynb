# Question no 3 answer

```python
import pandas as pd
records_dict={'Roll Number': ['20CSE29', '20CSE49', '20CSE36', '20CSE44'], 
                   'Name': ['Amelia', 'Sam', 'Dean', 'Jessica'], 
                   'Marks In Percentage': [97, 90, 70, 82], 
                   'Grade': ['A', 'A', 'C', 'B'], 
                   'Subject': ['Physics', 'Physics', 'Physics', 'Physics']}
print(records_dict)
print()
records=pd.Series(records_dict)
print(records)
print()
df=pd.DataFrame(records)
df['index']=records.index
df
```
